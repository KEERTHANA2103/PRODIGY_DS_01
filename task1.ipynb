#**Data Visualization**

##Visualization of Numerical Variables using Histogram

```python
cols = ['1960', '1970', '1980', '1990', '2000', '2022']
```

```python
for i in cols:
        fig = plt.figure(figsize=(6,4))
        plt.hist(df[i],color='blue',bins=10)
        plt.xlabel(i)
        plt.show()
```

```python
total_population=df[df['Indicator Code']=='SP.POP.TOTL']
```

```python
total_population_sort=total_population.sort_values(by='2003', ascending = False)
```

```python
total_population_top=total_population_sort.head(10)
total_population_top
```

```python
sns.histplot(total_population_top['2003'], kde=True, color='green')
plt.xlabel('Population in 2003', size=10)
plt.ylabel('Frequency', size=10)
plt.title('Histogram for Total Population in the year 2003', size=18)
```

```python
sns.histplot(total_population_top['1970'], kde=True, color='orange')
plt.xlabel('Population in 1970', size=10)
plt.ylabel('Frequency', size=10)
plt.title('Histogram for Total Population in the year 1970', size=18)
```

##Visualization of Categorical data using Bar graph

```python
max_10_population_year_2000 = df.sort_values(by='2000', ascending=False).head(10)
```

```python
plt.figure(figsize=(7,3))
plt.bar(max_10_population_year_2000['Country Name'], max_10_population_year_2000['2000'], color='purple')
plt.xlabel('Country Name')
plt.xticks(rotation=90)
plt.ylabel('Population')
plt.title('Top 10 Countries by Population in the Year 2000')
```

```python
min_10_population_year_2000 = df.sort_values(by='2000', ascending=False).tail(10)
```

```python
plt.figure(figsize=(7,3))
plt.bar(min_10_population_year_2000['Country Name'], min_10_population_year_2000['2000'], color='blue')
plt.xlabel('Country Name')
plt.xticks(rotation=90)
plt.ylabel('Population')
plt.title('Bottom 10 Countries by Population in the Year 2000')
```

```python
min_10_population_year_2022 = df.sort_values(by='2022', ascending=True).head(10)
```

```python
plt.figure(figsize=(7,3))
plt.bar(min_10_population_year_2022['Country Name'], min_10_population_year_2022['2022'], color='red')
plt.xlabel('Country Name')
plt.xticks(rotation=90)
plt.ylabel('Population')
plt.title('Top 10 Countries with minimum Population in the Year 2022')
```

```python
df_1=['Country Name','2020','2022']
df_2 = df[df_1].copy()
```

```python
df_2
```

```python
max_10_population_year_2020 = df_2.sort_values(by='2020', ascending=False).head(10)
```

```python
max_10_population_year_2020['Percentage_Increase'] = ((max_10_population_year_2020['2022'] - max_10_population_year_2020['2020']) / max_10_population_year_2020['2020']) * 100

# Plotting the bar graph
plt.figure(figsize=(10, 6))
plt.bar(max_10_population_year_2020['Country Name'], max_10_population_year_2020['Percentage_Increase'], color='skyblue')
plt.xlabel('Country')
plt.xticks(rotation=90)
plt.ylabel('Percentage Increase in Population')
plt.title('Percentage Increase in Population (2020 to 2022) of top 10 countries in Population')
plt.ylim(0, max(max_10_population_year_2020['Percentage_Increase']) + 10)  # Adjust ylim for better visualization
plt.grid(axis='y', linestyle='--', alpha=0.7)
```

```python
min_10_population_year_2020 = df_2.sort_values(by='2020', ascending=False).tail(10)
```

```python
min_10_population_year_2020['Percentage_Increase'] = ((min_10_population_year_2020['2022'] - min_10_population_year_2020['2020']) / min_10_population_year_2020['2020']) * 100

# Plotting the bar graph
plt.figure(figsize=(10, 6))
plt.bar(min_10_population_year_2020['Country Name'], min_10_population_year_2020['Percentage_Increase'], color='skyblue')
plt.xlabel('Country')
plt.xticks(rotation=90)
plt.ylabel('Percentage Increase in Population')
plt.title('Percentage Increase in Population (2020 to 2022) of top 10 countries in Population')
plt.ylim(0, max(min_10_population_year_2020['Percentage_Increase']) + 10)  # Adjust ylim for better visualization
plt.grid(axis='y', linestyle='--', alpha=0.7)
```
